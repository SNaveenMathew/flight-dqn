```python
import pandas as pd
import matplotlib.pyplot as plt

ac_data = pd.read_csv("../../dist_alt.csv")
```

```python
plt.scatter(ac_data["distance"], ac_data["altitude"], c = "lightgray")
plt.plot(ac_data["distance"][ac_data["id"] == "0D0418_3"], ac_data["altitude"][ac_data["id"] == "0D0418_3"], '-', c = "darkgreen", label = "Example Aircraft 1")
plt.plot(ac_data["distance"][ac_data["id"] == "A22702_14"], ac_data["altitude"][ac_data["id"] == "A22702_14"], '-', c = "red", label = "Example Aircraft 2")
plt.xlabel("Distance from JFK airport")
plt.ylabel("Altitude (ft.)")
plt.legend()
plt.xlim((0, 250000))
plt.show()
```

```python
ac_data = ac_data[ac_data['distance'] <= 250000]
ac_data = ac_data[(ac_data["id"] == "0D0418_3") | (ac_data["id"] == "A22702_14")]
ac2_data = ac_data[(ac_data["id"] == "A22702_14") & (ac_data["altitude"] <= 15000)]
```
